```python
def train_perceptron(x1, x2, y, w, alpha=0.1, max_epochs=100):
    def step(x):
        return 1 if x > 0 else 0

    for epoch in range(max_epochs):
        print(f"Epoch {epoch + 1}")
        weights_updated = False

        for i in range(len(y)):
            x = [1, x1[i], x2[i]]
            y_pred = step(w[0] * x[0] + w[1] * x[1] + w[2] * x[2])
            error = y[i] - y_pred

            if error != 0:
                weights_updated = True
                for j in range(len(w)):
                    w[j] += alpha * error * x[j]

            print(f"Input: {x[1:]}, Predicted: {y_pred}, Actual: {y[i]}, Weights: {w}")

        print("-" * 30)

        if not weights_updated:
            print("Convergence reached!")
            break

    print("Final weights:", w)
    return w
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(x1, x2, y, w, title='Perceptron Decision Boundary'):
    plt.figure(figsize=(6, 6))
    
    # Plot the data points
    for i in range(len(y)):
        if y[i] == 0:
            plt.scatter(x1[i], x2[i], color='red', marker='o', label='Class 0' if i == 0 else "")
        else:
            plt.scatter(x1[i], x2[i], color='blue', marker='x', label='Class 1' if i == 3 else "")

    # Plot the decision boundary
    x_values = np.linspace(-0.1, 1.1, 100)
    y_values = -(w[0] + w[1] * x_values) / w[2]
    plt.plot(x_values, y_values, label='Decision Boundary', color='green')

    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xlim(-0.1, 1.1)
    plt.ylim(-0.1, 1.1)
    plt.show()
```

```python
x1 = [0, 0, 1, 1]
x2 = [0, 1, 0, 1]
y = [0, 0, 0, 1]
w = [0.1, 0.2, 0.2]
train_perceptron(x1, x2, y, w, alpha=0.1)
plot_decision_boundary(x1, x2, y, w)
```
