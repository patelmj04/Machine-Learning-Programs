```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate custom dataset with predefined cluster centers
blob_centers = np.array([
    [0.2, 2.3],
    [-1.5, 2.3],
    [-2.8, 1.8],
    [-2.8, 2.8],
    [-2.8, 1.3]
])
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

X, y = make_blobs(n_samples=2000, centers=blob_centers,
                   cluster_std=blob_std, random_state=7)

# Apply original K-Means algorithm
kmeans = KMeans(n_clusters=5, init="random", n_init=1, algorithm="full", random_state=42)
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Function to plot clusters
def plot_clusters(X, labels, centroids):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=10, alpha=0.6)
    plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=200, label="Centroids")
    plt.xlabel("$x_1$", fontsize=14)
    plt.ylabel("$x_2$", fontsize=14, rotation=0)
    plt.title("K-Means Clustering (Original Algorithm)")
    plt.legend()

# Plot the clustered data
plt.figure(figsize=(8, 4))
plot_clusters(X, labels, centroids)
plt.show()
```
